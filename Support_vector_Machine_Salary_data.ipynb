```python
# combining Train and Test Data
combined_data = sal_train.append(sal_test)
combined_data.reset_index(inplace=True,drop=True)
combined_data
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]
```

```python
LE = LabelEncoder() #changing string data type to integer
for i in string_columns:
        sal_train[i]= LE.fit_transform(sal_train[i])
        sal_test[i]=LE.fit_transform(sal_test[i])
```

```python
sal_train.dtypes
```

```python
sal_test.dtypes
```

```python
#Model Building
X_train = sal_train.iloc[:,0:13]
Y_train = sal_train.iloc[:,13]
X_test = sal_test.iloc[:,0:13]
Y_test = sal_test.iloc[:,13]
```

```python
#Normalization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
X_train = norm_func(X_train)
X_test =  norm_func(X_test)
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
#Model Traing

svc_model=SVC()
```

```python
svc_model.fit(X_train,Y_train)
```

```python
#Model Testing
y_train_pred=svc_model.predict(X_train)
y_train_pred
```

```python
y_test_pred=svc_model.predict(X_test)
y_test_pred
```

```python
#Linear kernel

model_linear = SVC(kernel = "linear")
model_linear.fit(X_train,Y_train)
pred_test_linear = model_linear.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, pred_test_linear))
```

```python
#Polynomial kernel

model_poly = SVC(kernel = "poly")
model_poly.fit(X_train,Y_train)
pred_test_poly = model_poly.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, pred_test_poly))
```

```python
#RBF Kernel

model_rbf = SVC(kernel = "rbf")
model_rbf.fit(X_train,Y_train)
pred_test_rbf = model_rbf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, pred_test_rbf))
```

```python
#Sigmoid Kernel

model_sigmoid = SVC(kernel = "sigmoid")
model_sigmoid.fit(X_train,Y_train)
pred_test_sigmoid = model_sigmoid.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, pred_test_sigmoid))
```
