```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary_data=pd.read_csv('Salary_Data.csv')
salary_data
```

# EDA and Data Visualization

```python
salary_data.info()
```

```python
salary_data.describe().T
```

```python
plt.figure(figsize = (6, 4))
plt.tight_layout()
sns.distplot(salary_data.YearsExperience)
plt.show()

plt.figure(figsize = (6, 4))
plt.tight_layout()
sns.distplot(salary_data.Salary)
plt.show()
```

# Correlation Analysis

```python
salary_data.corr()
```

Correlation coefficient value r = 0.825997 indicates that there is a strong correlation between independent variable and dependent variable

```python
sns.regplot(x=salary_data['YearsExperience'],y=salary_data['Salary'])
```

### Model Building

```python
model_slr=smf.ols("Salary~YearsExperience",data=salary_data).fit()
```

# Getting important parameters from model

```python
# Coefficients Beta0 and Beta1
Beta0, Beta1 = model_slr.params
print("Intercept, Beta0: ", Beta0)
print("YearsExperience, Beta1: ", Beta1)

# Getting tvalue and pvalue
tvalue, pvalue = (model_slr.tvalues, model_slr.pvalues)
print("tvalue: ", tvalue)
print("pvalue: ", pvalue)

# Getting R squared values
r_sq, r_sq_adj = (model_slr.rsquared, model_slr.rsquared_adj)
print("R2: ", r_sq)
print("R2_Adj: ", r_sq_adj)
```

# Checking predictions of Model by equation

```python
predicted_salary = model_slr.predict(salary_data.YearsExperience)
pred_df = pd.DataFrame({'Salary_Actual' : salary_data.Salary, 'Salary_Predicted' : predicted_salary})
pred_df
```

```python
# Plotting Actual Vs. Predicted Values
plt.scatter(salary_data.YearsExperience, salary_data.Salary,  color='green')
plt.plot(salary_data.YearsExperience, predicted_salary, color='red', linewidth=2)
plt.title('Actual Vs. Predicted Values')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Calculating Errors

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(salary_data.Salary, predicted_salary))  

print('Mean Squared Error:', metrics.mean_squared_error(salary_data.Salary, predicted_salary))  

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(salary_data.Salary, predicted_salary)))

print("R^2 Score :          ", metrics.r2_score(salary_data.Salary, predicted_salary))
```

# Checking Normality of Residuals Using the Anderson-Darling test for normal distribution

```python
from statsmodels.stats.diagnostic import normal_ad
residuals = salary_data.Salary - predicted_salary
p_value_thresh=0.05
# Performing the test on the residuals
p_value = normal_ad(residuals)[1]
print('p-value from the test: below 0.05 generally means non-normal:', p_value)

# Reporting the normality of the residuals
if p_value < p_value_thresh:
    print('Residuals are not normally distributed')
else:
    print('Residuals are normally distributed')

# Plotting the residuals distribution
plt.subplots(figsize=(6, 4))
plt.title('Distribution of Residuals')
sns.distplot(residuals)
plt.show()
```
