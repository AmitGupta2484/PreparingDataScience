# Dropping the month and day columns

```python
df.drop(["month","day"],axis=1,inplace =True)
```

```python
df
```

```python
predictors = df.iloc[:,0:28] # all rows and 1 to 28th columns
target = df.iloc[:,28] # all rows and 29th column
```

```python
predictors
```

```python
target
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
##Normalising the data as there is scale difference
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(predictors) #Normalised predictors data
fires
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Fit (train) the Support Vector Machine classifier using kernel poly
svm_clf = SVC(kernel = "poly")
svm_model = svm_clf.fit(x_train, y_train)
svm_prediction = svm_clf.predict(x_test)
svm_train_pred = svm_clf.predict(x_train)

print("Accuracy {0:.2f}%".format(100*accuracy_score(svm_train_pred, y_train)))
print("Accuracy {0:.2f}%".format(100*accuracy_score(svm_prediction, y_test)))


# Display the Confusion Matrix and Classification Report 
print(confusion_matrix(svm_prediction, y_test))
print(classification_report(svm_prediction, y_test))
```

```python
# Fit (train) the Support Vector Machine classifier using kernel liniear
svm_clf = SVC(kernel = "linear")
svm_model = svm_clf.fit(x_train, y_train)
svm_prediction = svm_clf.predict(x_test)
svm_train_pred = svm_clf.predict(x_train)

print("Accuracy {0:.2f}%".format(100*accuracy_score(svm_train_pred, y_train)))
print("Accuracy {0:.2f}%".format(100*accuracy_score(svm_prediction, y_test)))


# Display the Confusion Matrix and Classification Report 
print(confusion_matrix(svm_prediction, y_test))
print(classification_report(svm_prediction, y_test))
```
